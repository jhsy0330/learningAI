$$ y = 2 * X + 4 $$

```python
import torch

# 定义模型参数
w = torch.tensor(1.5, requires_grad=True)  # 权重
b = torch.tensor(0.0, requires_grad=True)  # 偏置

# 定义损失函数
def square_loss(y_hat, y):
    return (y_hat - y) ** 2 / 2

# 定义模型
def net(X):
    return X * w + b

# 定义优化器（SGD）
def sgd(params, lr, batch_size):
    for param in params:
        param.data -= lr * param.grad / batch_size
        param.grad.zero_()

# 定义真实数据
X_truth = torch.tensor(range(10), dtype=torch.float32)
B_truth = 4
Y_truth = 2 * X_truth + B_truth

# 定义训练函数
def train(net, X, Y, loss, num_epochs, lr):
    for epoch in range(num_epochs):
        #清除梯度
        if w.grad is not None:
            w.grad.zero_()
        if b.grad is not None:
            b.grad.zero_()

        # for i in range(0, len(X), batch_size):
        #     x_batch = X[i:i + batch_size]
        #     y_batch = Y[i:i + batch_size]
        #     l = loss(net(x_batch), y_batch)
        #     l.backward()
        #     sgd([w, b], lr, batch_size)

        for x, y in zip(X, Y):
            l = loss(net(x), y)
            l.backward()
            sgd([w, b], lr, 1)

        with torch.no_grad():
            l = loss(net(X), Y)
            print(f'epoch {epoch + 1}, loss {l.mean()}')

# 训练模型
train(net, X_truth, Y_truth, square_loss, 1000, 0.01)

# 打印训练后的参数
print(f'训练后的权重 w: {w.item()}')
print(f'训练后的偏置 b: {b.item()}')
```
